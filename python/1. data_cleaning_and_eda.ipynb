# Splitting Numerical and Categorical Features

```python
numeric_features = []
categorical_features = []

for column in df.columns:
    if df[column].dtype == "str":
        categorical_features.append(column)
    else:
        numeric_features.append(column)
```

```python
categorical_features
```

# Data Cleaning

```python
df_copy = df.copy()
```

## Cleaning - Categorical Fetures

```python
df[categorical_features].info()
```

```python
for column in categorical_features:
    print(df[column].unique())
    print()
```

- Since the `categorical` features do not contain any missing values or duplicate values, no data cleaning was required for the categorical features.


---

## Cleaning - Numerical Features

```python
df[numeric_features].describe()
```

```python
df[numeric_features].isnull().sum()
```

### Null Imputation - `rating_given`

```python
df["rating_given"].value_counts()
```

```python
rating_given_mode = df["rating_given"].mode()[0]

df["rating_given"] = df["rating_given"].fillna(rating_given_mode)
```

- Missing values in the `rating_given` column were filled using the `mode` value.
